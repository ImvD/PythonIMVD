
3. Factorial de un número:
- Escribe un programa que calcule el factorial de un número dado.

````python
def factorial(numero):
    resultado = 1
    for i in range(1, numero + 1):
        resultado *= i
    return resultado

try:
    numero_inicial = int(input("Dame un número: "))
    if numero_inicial < 0:
        print("Por favor, introduce un número no negativo.")
    else:
        print(f"El factorial de {numero_inicial} es {factorial(numero_inicial)}")
except ValueError:
    print("Por favor, introduce un número válido.")
````

````python
# Función para calcular el factorial de un número
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

# Calcular el factorial dado
valor = int(input("Indica el número del que quieres obtener el factorial: "))
resultado =  factorial(valor)
print(f"El factorial de {valor} es:{resultado}")
````


5. Números primos en un rango:
- Escribe un programa que encuentre todos los números primos en un rango dado.

````python
def generar_lista(a, b):
    return [i for i in range(a, b + 1)]

def es_primo(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def comprobar_primos(lista):
    return [i for i in lista if es_primo(i)]

try:
    a = int(input("Indica el comienzo del rango: "))
    b = int(input("Indica el final del rango: "))
    if a >= b:
        print("El valor inicial debe ser menor al valor final.")
    else:
        lista_final = generar_lista(a, b)
        lista_primos = comprobar_primos(lista_final)
        print(lista_primos)
except ValueError:
    print("Por favor, introduce números enteros válidos.")
````


9. **Conversor de Temperatura**: Escribe un programa que convierta temperaturas de Celsius a Fahrenheit y viceversa.

````python
def conversor_temperatura(t,tip):
    if tip.lower() == "fahrenheit":
        t = (t - 32) / 1.8
        tip = "Celsius"
    elif tip.lower() == "celsius":
        t = (t * 1.8) + 32
        tip = "Fahrenheit"
    else:
        print("Por favor, introduce un tipo de temperatura al que quieras convertir correcto")
    return t, tip

temperatura = int(input("Introduce la temperatura a convertir: "))
tipo = str(input("Introduce el tipo de temperatura (Celsius o Fahrenheit)"))

temperatura_conv, tipo_conv = conversor_temperatura(temperatura,tipo)

print(f"Los {temperatura}º {tipo} equivalen a {temperatura_conv}º {tipo_conv}")
````

````python
def calculadora(n1,n2,func):
    num1 = n1
    num2 = n2
    if func.strip().lower() == "sumar":
        return n1 + n2
    elif func.strip().lower() == "restar":
        return n1 - n2
    elif func.strip().lower() == "multiplicar":
        return n1 * n2
    elif func.strip().lower() == "dividir":
        return n1 / n2
    else:
        print("Introduce una función válida")

numero1,numero2,funcion = int(input("Introduce el valor 1: ")),int(input("Introduce el valor 2: ")),str(input("Introduce la función (sumar,restar,multiplicar o dividir): "))

print(f"El resultado de {funcion} {numero1} y {numero2} es: {calculadora(numero1,numero2,funcion)}")
````

13. **Generador de Contraseñas**: Escribe un programa que genere una contraseña aleatoria de una longitud dada.

````python
import random
import string

def generador_contraseña(n):
    if n < 4:
        raise ValueError("La longitud de la contraseña debe ser al menos 4 para incluir todos los tipos de caracteres.")
    
    caracteres = string.ascii_letters + string.digits + string.punctuation
    contraseña = [
        random.choice(string.ascii_uppercase),
        random.choice(string.ascii_lowercase),
        random.choice(string.digits),
        random.choice(string.punctuation)
    ]
    
    contraseña += random.choices(caracteres, k=n-4)
    random.shuffle(contraseña)
    
    return ''.join(contraseña)

try:
    longitud = int(input("Introduce la longitud de la contraseña: "))
    contraseña = generador_contraseña(longitud)
    print(f"La contraseña generada es: {contraseña}")
except ValueError as e:
    print(e)
````


15. **Juego de Piedra, Papel o Tijera**: Escribe un programa que permita al usuario jugar contra la computadora.

````python
import random

def juego_ppt():
    elementos = ["Piedra", "Papel", "Tijera"]
    print("Hola, bienvenido al juego de Piedra, Papel o Tijera, vamos a jugar al mejor de 3 partidas. \n Comenzamos ")
    c1, c2 = 0,0
    while max(c1,c2) < 3:
        u_choice = str(input("Elige entre Piedra, Papel o Tijera: "))
        c_choice = random.choice(elementos)
        if (u_choice.lower() == c_choice.lower()):
            print(f"Habéis empatado: \n User: {u_choice} , Computer: {c_choice}")
        elif( (u_choice.lower() == "tijera" and c_choice.lower() =="papel")
             or (u_choice.lower() == "piedra" and c_choice.lower() =="tijera")
             or (u_choice.lower() == "papel" and c_choice.lower() =="piedra")):
            print(f"Has ganado!!: \n User: {u_choice} , Computer: {c_choice}")
            c1 +=1
            print(f"U:{c1} C:{c2}")
        else:
            print(f"He ganado yo!!: \n User: {u_choice} , Computer: {c_choice}")              
            c2 +=1          
            print(f"U:{c1} C:{c2}")
    if c1 == 3:
        print("Has ganado la partida!!")
    else:
        print("He ganado yo ya partida!!")
    
juego_ppt()
````

````python
from collections import Counter
vocales = list("aeiou")
def devuelve_vocales(frase):
    return Counter(letra for letra in frase.lower() if letra in vocales)

frase_entrada = input(str("Introduce la frase para la que comprobar las vocales"))
devuelve_vocales(frase_entrada)
````


6. **Palabras en mayúsculas**: Convierte todas las palabras de una lista a mayúsculas.

````python
def conversor_mayus(l):
    lista_mayus = [i.upper() for i in l]
    return [j for j in lista_mayus if j.strip()]

lista = list("hola que tal")

conversor_mayus(lista)
````


7. **Suma de pares**: Calcula la suma de todos los números pares del 1 al 100.

````python
def suma_pares():
    return sum(i for i in range (1,101) if i%2 == 0)
print(suma_pares())
````


8. **Números impares al cuadrado**: Genera una lista de los cuadrados de los números impares del 1 al 20.

````python
def cuadrado_impar():
    return [i**2 for i in range(1,21) if (i%3==0 or i%5==0)]

print(cuadrado_impar())
````


9. **Palabras con más de 3 letras**: Filtra las palabras de una lista que tienen más de 3 letras.

````python
def filtro_3(l):
    return [i for i in l if len(i)>3 ]

print(filtro_3(("Hola como estas esto es una prueba").split()))
````

````python
def divisibles_3_5():
    return [i for i in range (1,101) if (i%3==0 and i%5==0)]
print(divisibles_3_5())
````
