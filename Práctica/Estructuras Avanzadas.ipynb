### **Diccionarios**

1. Crea un diccionario vacío y añade tres pares clave-valor.

````python
Diccionario1 = {
    "Nombre" : "Iván",
    "Edad": 30,
    "Profesión": "Pintor"
}
for clave,valor in Diccionario1.items():
    print(f"{clave}, {valor}")
````


2. Dado un diccionario, imprime todas sus claves.

````python
print(Diccionario1.keys())
````


3. Dado un diccionario, imprime todos sus valores.

````python
print(Diccionario1.values())
````


4. Dado un diccionario, imprime todos los pares clave-valor.

````python
for clave1, valor1 in Diccionario1.items():
    print(f"{clave1} y {valor1}")
````


5. Combina dos diccionarios en uno solo.

````python
Diccionario2 = {
    "Nombre":"Arantxa",
    "Edad":"27",
    "Profesión":"Pintora"    
}

Diccionario1.update(Diccionario2)

print(Diccionario1.items())
````


6. Dado un diccionario, elimina una clave específica.

````python
Diccionario1.pop("Edad")

print(Diccionario1.items())
````


7. Dado un diccionario, verifica si una clave existe.

````python
Diccionario1.get("Nombre")
````


8. Crea un diccionario donde las claves sean números del 1 al 5 y los valores sean sus cuadrados.

````python
Diccionario_cuadrados = {x:x**2 for x in range (1,6)}

Diccionario_cuadrados.values()
````


9. Dado un diccionario, encuentra la clave con el valor máximo.

````python
clave = [clave for clave,valor in Diccionario_cuadrados.items() if valor == max(Diccionario_cuadrados.values())]

print(clave) 
````


10. Dado un diccionario, invierte las claves y los valores (los valores pasan a ser claves y viceversa).

````python
diccionario_inverso = {valor:clave for clave, valor in Diccionario1.items()}

print(diccionario_inverso.items())
````

### **Conjuntos**
11. Crea un conjunto con los números del 1 al 10.

````python
conjunto1 = {x for x in range (1,11)}
conjunto2 = {x for x in range (11,21)}
print(conjunto1)
print(conjunto2)
````


12. Dados dos conjuntos, encuentra su unión.

````python
print(conjunto1.union(conjunto2))
print(len(conjunto1.union(conjunto2)))
````


13. Dados dos conjuntos, encuentra su intersección.

````python
print({x for x in conjunto1 & conjunto2})
````


14. Dados dos conjuntos, encuentra la diferencia entre el primero y el segundo.

````python
print({x for x in conjunto2 - conjunto1})
````


15. Dado un conjunto, verifica si es subconjunto de otro.

````python
print(conjunto1.issubset(conjunto2))
````


16. Elimina un elemento específico de un conjunto.

````python
conjunto1.remove(1)
print(conjunto1)
````


17. Dado un conjunto, convierte sus elementos en una lista.

````python
conjunto3 = list(conjunto1)
print(conjunto3)
````


18. Crea un conjunto a partir de una lista, eliminando duplicados.

````python
lista = [1,2,1,1,3,44,5,6,8,6,4,3]
print(lista)
conjunto_l = set(lista)
print(conjunto_l)
````


19. Dado un conjunto, añade varios elementos a la vez.

````python
conjunto_l.update([7,9,10])
print(conjunto_l)
````


20. Dado un conjunto, elimina todos sus elementos.

````python
conjunto_l.clear()
print(conjunto_l)
````
