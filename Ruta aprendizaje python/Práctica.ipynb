Escribe un programa que pida al usuario un n√∫mero y determine si es positivo, negativo o cero. Muestra un mensaje apropiado para cada caso.

---

````python
def pos_neg_cer(numero):
    if numero < 0:
        print(f"El n√∫mero {numero} es negativo")
    elif numero > 0:
        print(f"El n√∫mero {numero} es positivo")
    else:
        print(f"El n√∫mero {numero} es cero")
        
numero1 = int(input("Introduce el n√∫mero 1: "))
pos_neg_cer(numero1)
````

### üìå **Ejercicio 3: Calculadora simple**
**Enunciado:**  
Crea una calculadora que pida dos n√∫meros y una operaci√≥n (+, -, *, /). Realiza la operaci√≥n correspondiente y muestra el resultado. Maneja el caso especial de divisi√≥n por cero.

---

````python
def calculadora(num1,num2,op):
    resultado = 0
    opc = list(op.strip().capitalize())
    
    if opc[0] == "S":
        resultado = num1 + num2
    else:
        resultado = 0
    return resultado,opc
    
    
numero1 = int(input("Introduzca el primer n√∫mero: "))
numero2 = int(input("Introduzca el segundo n√∫mero: "))
operacion = str(input("Introduce la operaci√≥n que quieres realizar"))

resul,operation = calculadora(numero1,numero2,operacion)
print(f"El resultado de la {operacion} entre {numero1} y {numero2} es: {resul}")
````
